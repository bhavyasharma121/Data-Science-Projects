Heaven Furnitures (HF) sells furniture like sofas, beds and tables. It is observed that 25% of their customers complain about the furniture purchased by them for many reasons. On Tuesday, 20 customers purchased furniture products from HF

a. Calculate the probability that exactly 3 customers will complain about the purchased products.

```python
p = 0.25
n = 20
x = 3
stats.binom.pmf(x, n, p)
```

* PMF stands for Probability Mass Function
* PMF is calculated for that Instance.
* CDF Stands for Cumulative Density Function
* CDF returns 1 in the End as it is a cumulative probability that adds to 1

```python
# Distribution of PMF
x = np.arange(0, 20)
plt.plot(stats.binom.pmf(x, n, p), "o-", color = "red", )
plt.title("Binomial PMF")
plt.xlabel("No of Customers")
plt.ylabel("PMF P(X)")
plt.show()
```

```python
# Distribution of CDF
x = np.arange(0, 20)
plt.plot(stats.binom.cdf(x, n, p), "o-", color = "red", )
plt.title("Binomial CDF")
plt.xlabel("No of Customers")
plt.ylabel("CDF P(X)")
plt.show()
```

## About Standard Error

The standard error tells you how accurate the **mean of any given sample** from that population is likely to be compared to the true population mean. 

When the **standard error increases, i.e. the means are more spread out**, it becomes more likely that **any given mean 
is an inaccurate representation of the true population mean.**

**Standard error decreases when sample size increases** – as the sample size gets closer to the true size of the population, the sample means cluster more and more around the true population mean.

If you take enough samples from a population, the means will be arranged into a distribution around the true population mean. The standard deviation of this distribution, i.e. the standard deviation of sample means, is called the standard error.

### Sampling from a Normally Distributed Population

When the samples are drawn from the Normally Distributed Population with the mean is **mew** and Standard Deviation is equal to **sigma**

* The Mean (Xbar) rep as **Mew of Xbar** is equal to the pop mean Mew.

* The Standard Deviation of Xbar rep as **sigma of xbar** is equal to sigma/root(n). 

### Question on Finding the Standard Error
The mean wage per hour for all 5000 employees who work at a large company is 27.50, and the standard deviation is 3.70. Let x be the mean wage per hour for a random sample of certain employees selected from this company.

Find the mean and standard deviation of x for a sample size of 30, 75 and 200

```python
# Population Mean, Median, SD ~ Population Parameter...
# Sample Data - Mean, Median....~ Sample Statistic...
mew = 27.50
sigma = 3.70
size = [30, 75, 200]
# Since we know that the mean of samples is equal to Pop Mean
# thus, these samples can presume to have the Pop Mean...
# Therefore, the xbar of all the samples is assumed to be mew...

def standarderror(sd, n):
    return(sd/np.sqrt(n))

for i in size:
    print("S.E. for ", i, " Sample is ", standarderror(sigma, i))
```

Note: The difference between the **Sample Mean and the Population Mean** is Called **Sampling Error** (xbar - mew)

Assume that the weights of all packages of a certain brand of cookies are approximately normally distributed with a mean of 32 ounces and a standard deviation of .3 ounce.Find the probability that the mean weight, x, of a random sample of 20 packages of this brand of cookies will be between 31.8 and 31.9 ounces.

```python
popmean = 32
popsd = 0.3
n = 20
stats.norm.cdf(31.9, popmean, 
               popsd/np.sqrt(n)) - stats.norm.cdf(31.8, popmean, 
                                                  popsd/np.sqrt(n))
```

### Understanding Empirical Rule

* From CLT, it is clear that for large Samples (30 or more), the sampling distribution of xbar is approximately normal with the mean mew and standard deviation being sigma/root(n).

* If we take large samples from a population and calculate the mean of all the samples, then 68.26% of the sample means will be within 1 standard deviaton of the pop mean.

* If we take large samples from a population and calculate the mean of all the samples, then 95.44% of the sample means will be within 2 standard deviaton of the pop mean.

* If we take large samples from a population and calculate the mean of all the samples, then 99.74% of the sample means will be within 3 standard deviaton of the pop mean.

```python
# Calculate Empirical Rule
stats.norm.cdf(1, 
               loc = 0, 
               scale = 1) - stats.norm.cdf(-1, 
                                           loc = 0, 
                                           scale = 1)
```
